```python
import bs4 as bs
import urllib.request #an auxiliary lib to request the http data
sauce = urllib.request.urlopen("https://en.wikipedia.org/wiki/Wheel").read()
soup = bs.BeautifulSoup(sauce, 'lxml')

#Some intuitive applications:

print(soup.title)
print("************")
print(soup.title.name)
print("************")
print(soup.title.string)
print("************")
print(soup.title.text)
print("************")
print(soup.p) #first paragraf element
print("************")
par = soup.find_all('p') #find all paragraphs ('p')
```

###### This would print EVERYTHING written in the http

```python
# print(soup.get_text()) 
```

###### This will print only the text of each paragraph using for loop, looks nicer to work with it.

```python
for paragraph in soup.find_all('p'):
    print(paragraph.text) # '.string' would include child tags, wouldnt look good
```

###### You can also look for all the links in the http

##### Maybe to create a page rank :)

```python
for links in soup.find_all('a'):
    print(links.get('href'))
```
